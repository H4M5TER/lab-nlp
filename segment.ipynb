```python
from icecream import ic

ic.configureOutput(argToStringFunction=str)  # fix for HanLP JClass

processors = []
```

```python
import jieba

processors.append(
  ("jieba", lambda x: jieba.lcut(x)),
)
```

```python
from pyhanlp import HanLP

processors.append(
  ("hanlp1.x", lambda x: HanLP.segment(x)),
)
```

```python
import hanlp
from hanlp.pretrained import tok

hanlp_coarse = hanlp.load(tok.COARSE_ELECTRA_SMALL_ZH)
processors.append(
  ("hanlp2coarse", lambda x: hanlp_coarse(x)),
)
hanlp_fine = hanlp.load(tok.FINE_ELECTRA_SMALL_ZH)
processors.append(
  ("hanlp2fine", lambda x: hanlp_fine(x)),
)

# hanlp_minilm = hanlp.load(tok.UD_TOK_MMINILMV2L12)  # bad quality
# processors.append(
#   ("hanlp2minilm", lambda x: hanlp_minilm(x)),
# )
```

```python
from snownlp import SnowNLP

processors.append(
  ("snownlp", lambda x: SnowNLP(x).words),
)
```

```python
from pkuseg import pkuseg

pkuseg_default = pkuseg()
pkuseg_default2 = pkuseg("D:/nlp/models/pkuseg/default_v2")

processors.append(
  ("pkuseg1", lambda x: pkuseg_default.cut(x)),
)
processors.append(
  ("pkuseg2", lambda x: pkuseg_default2.cut(x)),
)
```

```python
from thulac import thulac

thulac_instance = thulac(seg_only=True)
processors.append(
  ("thulac", lambda x: thulac_instance.cut(x, text=True).split(" ")),
)
```

```python
from ltp import LTP

ltp = LTP("LTP/base1")
ltp.to("cuda")

processors.append(
  ("LTP", lambda x: ltp.pipeline(x, tasks=["cws"])),
)
```
